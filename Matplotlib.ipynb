# Format Strings fmt
You can also use the shortcut string notation parameter to specify the marker.

This parameter is also called fmt, and is written with this syntax:

marker|line|color

```python
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, 'o:r')  # Marker, line, color
plt.show()
```

# Marker Size
You can use the keyword argument markersize or the shorter version, ms to set the size of the markers:

```python
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker = 'o', ms=20)  # ms  means marker size
plt.show()
```

# Marker Color
You can use the keyword argument markeredgecolor or the shorter mec to set the color of the edge of the markers:

```python
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker = 'o', ms = 20, mec = 'r')
plt.show()
```

You can use the keyword argument markerfacecolor or the shorter mfc to set the color inside the edge of the markers:

```python
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker = 'o', ms = 20, mfc = 'r')
plt.show()
```

# Set Font Properties for Title and Labels
You can use the fontdict parameter in xlabel(), ylabel(), and title() to set font properties for the title and labels.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

font1 = {'family':'sans-serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}

plt.title("Sports Watch Data", fontdict = font1)
plt.xlabel("Average Pulse", fontdict = font2)
plt.ylabel("Calorie Burnage", fontdict = font2)

plt.plot(x, y)
plt.show()
```

# The subplot() Function
* The subplot() function takes three arguments that describes the layout of the figure.

*  The layout is organized in rows and columns, which are represented by the first and second argument.

* The third argument represents the index of the current plot.

* plt.subplot(1, 2, 1)
the figure has 1 row, 2 columns, and this plot is the first plot.

*
plt.subplot(1, 2, 2)
the figure has 1 row, 2 columns, and this plot is the second plot.

```python
import matplotlib.pyplot as plt
import numpy as np

#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(3, 1, 1)
plt.plot(x,y)

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(3, 1, 2)
plt.plot(x,y)

#plot 3:
x = np.array([0, 1, 2, 3])
y = np.array([10, 5, 30, 40])

plt.subplot(3, 1, 3)
plt.plot(x,y)



plt.show()
```

# Multiple Bar In Frame

```python
import matplotlib.pyplot as plt
import numpy as np

# set width of bar
barwidth = 0.25
fig = plt.subplots(figsize=(10,4))

# set height of bar
IT = [12, 30, 1, 8, 22]
ECE = [28, 6, 16, 5, 10]
CSE = [29, 3, 24, 25, 17]

# Bar Positions
br1 = np.arange(len(IT))
br2 = [x + barwidth for x in br1]
br3 = [x + barwidth for x in br2]

# Plotting the Bars
plt.bar(br1, IT, color='r', width=barwidth, edgecolor='grey', label='IT')
plt.bar(br2, ECE, color='g', width=barwidth, edgecolor='grey', label='ECE')
plt.bar(br3, CSE, color='b', width=barwidth, edgecolor='grey', label='CSE')

# Adding Labels
plt.xlabel('Branch', fontweight='bold', fontsize=15)
plt.ylabel('Students Passed', fontweight='bold', fontsize=15)

# Places custom tick marks on the x-axis, centering them under the grouped bars. Labels (2015 to 2019) represent years.
plt.xticks([r + barwidth for r in range(len(IT))],
        ['2015','2016','2017','2018','2019'])

plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

n=5

boys=(20,35,30,35,27)
girls=(25,32,34,20,25)
boystd=(2,3,4,1,2)
girlstd=(3,5,2,3,3)
ind=np.arange(n)
width=0.35

fig=plt.subplots(figsize=(10,7))
p1=plt.bar(ind, boys, width, yerr=boystd)
p2=plt.bar(ind, girls, width, bottom=boys, yerr=girlstd)

plt.ylabel('Contribution')
plt.title('Contribution by the teams')
plt.xticks(ind,('T1','T2','T3','T4','T5'))
plt.yticks(np.arange(0,81,10))
plt.legend((p1[0],p2[0]),('Boys','Girls'))

plt.show()
```

```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5]
y = [10,15,13,18,16]

plt.plot(x,y, marker='o')

plt.annotate('highest point', xy=(4,18), xytext=(3,20), arrowprops=dict(arrowstyle='->',color='red'))

for i,j in zip(x,y):
  plt.text(i,j+0.5, f'{j}',ha='center',color='green')

plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('annotated line chart')

plt.show()
```

* plt.annotate: This function is used to add annotations to a plot. Annotations are text labels that can point to specific data points on the graph.

* 'highest point': This is the text of the annotation. It will appear at the location specified by xytext.

* xy=(4,18): This specifies the coordinates of the point to be annotated. The arrow will point to this location on the plot.

* xytext=(3,20): This specifies the coordinates where the annotation text ('highest point') will appear on the plot. It's positioned away from the actual point (xy) to avoid overlapping.

* arrowprops=dict(arrowstyle='->', color='red'):

* arrowprops: A dictionary defining the properties of the arrow that connects the text to the point.
* arrowstyle='->': Specifies that the arrow should have a simple arrowhead.
* color='red': Sets the color of the arrow to red.

```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5]
y = [10,15,13,17,16]

plt.plot(x,y,marker='o',color='b',linestyle='-',label='sample line')

plt.annotate('highest point',xy=(4,17),xytext=(3,20))

# annotate each data points with its value
for i,j in zip(x,y):
  plt.text(i,j+0.5, f'({i},{j})',ha='center',color='green')

plt.title('line chart with annotation',loc='left')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.legend()
plt.show()
```

# over laped stacked graph

```python
import matplotlib.pyplot as plt
import numpy as np

# Create figure and 3D axes
fig = plt.figure()
ax = plt.axes(projection='3d')

# Define data
x = np.linspace(-5, 5, 50)
y = np.linspace(-5, 5, 50)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

# Plot the surface
ax.plot_surface(X, Y, Z, cmap='viridis')

# Add labels
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("/content/company-sales.csv")


x = df['month_number'].tolist()
y = df['facecream'].tolist()
z = df['facewash'].tolist()

plt.bar([a-0.25 for a in x], y, color='b',width=0.25, label = 'face cream sales data', align='edge')
plt.bar([a+0.25 for a in x], z, color='orange',width= -0.25, label = 'face wash sales data', align='edge')

plt.xlabel('Month number')
plt.ylabel('Sales unit in number')
plt.title('face cream and face wash sales data ')

plt.grid(linestyle='dashed',linewidth=1)
plt.xticks(x)
plt.legend(loc='upper left')
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

df=pd.read_csv("/content/company-sales.csv")

a = df['month_number'].tolist()
b = df['facecream'].tolist()
c = df['facewash'].tolist()
d = df['toothpaste'].tolist()
e = df['bathingsoap'].tolist()
f = df['shampoo'].tolist()
g = df['moisturizer'].tolist()

plt.plot(a,b,color='red',marker='o',label = 'Face creame sales data')
plt.plot(a,c,color='green',marker='o',label = 'Face wash sales data')
plt.plot(a,d,color='blue',marker='o',label = 'toothpaste sales data')
plt.plot(a,e,color='purple',marker='o',label = 'shampoo sales data')
plt.plot(a,f,color='brown',marker='o',label = 'moisturizer sales data')

plt.xlabel('month number')
plt.ylabel('sales unit in number')
plt.title('sales data')
plt.xticks(a)
plt.legend()
plt.show()
```

```python
x = np.linspace(-10,10,100)
y = x**2
plt.subplot(2,2,1)
plt.plot(x,y)
plt.title('line plot')

x = [-10,0,10,12,20,102,3,9,2]
y = [100,0,100,45,2,5,6,2,56]
plt.subplot(2,2,2)
plt.scatter(x,y)
plt.title('scatter plot')

x = ['A','B','C']
y = [3,7,2]
plt.subplot(2,2,3)
plt.bar(x,y)
plt.title('bar plot')

x = np.random.randn(1000)
plt.subplot(2,2,4)
plt.hist(x,bins=50)
plt.title('histogram')

plt.tight_layout()
plt.show()
```

```python
x = np.linspace(-2,4,100)
y = -x**2+4*x

plt.fill_between(x,y,color="lightblue",alpha=0.6)
plt.plot(x,y,label='-x**2+4*x')

plt.legend()
plt.show()
```
